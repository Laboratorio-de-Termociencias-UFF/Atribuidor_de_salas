Mesmo caso para a sala 446: ficará 446-D e 446-E, já que também tem duas com mesmo nome nos dois blocos<br>
Modifiquei isso na planilha Horário_new<br><br>
Implementar critério 1: capacidade da sala deve ser maior ou igual ao preechimento médio das disciplinas<b>(Done)</b><br>
Implementar critério 2: primeira verificação deve ser feita pelo preferência inerente àquela sala<b>(Peding)</b><br>
Implementar critério 3: verificar se a sala também está livre no segundo dia de aula, pois critério não foi considerado, e pode ter o risco de ser sobreposta, porque ele só está verificando se ela está livre no primeiro dia de aula da disciplina<b>(Peding)</b>

```python
salas = df_tempo['Sala'].unique()
disciplinas_fora = []
for i in range(df_aulas.shape[0]):
    # 
    duration = int(df_aulas.iloc[i]['Duracao'])
    begin = df_aulas.iloc[i]['COMEÇO'].hour
    # pegando o primeiro dia da semana que tem aula
    days = []
    for num in range(8, 14):
        if df_aulas.iloc[i][df_aulas.columns[num]] == 1: # Aqui ele está pegando todos os dias da semana que tem aula
            days.append(df_aulas.columns[num])
    print(days)
    # Verificando se a sala está vazia, considerando as condições. Se sim, então atribuir sala à disciplina
    try:
        # Neste bloco será verificado o departamento das disciplinas compatíveis com o departamento das salas
        for n in range(len(salas)):
            print(df_tempo[df_tempo['Sala'] == salas[n]]['Dept'].unique()[0])
            print(df_aulas.iloc[i]['Departamento'])
            if df_tempo[(df_tempo['Dia da semana'] == days[0])&(df_tempo['Horário'] == begin)&(df_tempo['Sala'] == salas[n])]['Disciplina'].item() == "" and df_tempo[df_tempo['Sala'] == salas[n]]['Capac.'].iloc[0] >= df_aulas.iloc[i]['Preechimento médio'] and df_tempo[df_tempo['Sala'] == salas[n]]['Dept'].unique()[0] == df_aulas.iloc[i]['Departamento']:
                print("begin: ", begin)
                for d in days:
                    begin1 = begin
                    f = 0
                    while f != duration:
                        index = df_tempo[(df_tempo['Dia da semana'] == d)&(df_tempo['Horário'] == begin1)&(df_tempo['Sala'] == salas[n])].index
                        df_tempo.iloc[index, 7] = df_aulas.loc[i]['Disciplina']
                        df_tempo.iloc[index, 6] = df_aulas.loc[i]['Departamento']
                        df_tempo.iloc[index, 8] = df_aulas.loc[i]['Professor']
#                         print("begin1: ", begin1)
#                         print(index)
                        begin1 += 1
                        f += 1
                df_aulas.iloc[i, 14] = salas[n] #Acrescentar sala para indicar que determinada disciplina agora tem uma sala
                print('\033[1;34mA sala', salas[n], "no horário", begin1, "da", d, 'teve a disciplina', df_aulas.loc[i]['Disciplina'],'atribuída\033[m')
                break
            else:
                print('\033[0;31mA sala ', salas[n], "no horário", begin, "do dia ", days[0], 'tá ocupada', 'ou não teve capacidade\033[m')
                if n == (len(salas) - 1):
                    disciplinas_fora.append(df_aulas.loc[i]['Disciplina'])
                continue
        
        
    except:
        print("Esse filha da puta aqui que atrapalhou", salas[n], begin, days[0])
        raise
    print("Foi")
    days.clear()
print('Disciplinas que ficaram de fora: ', disciplinas_fora)
```
